```python
#The best math results are at least 80% of the *maximum possible score* for math.Save your results in a pandas DataFrame called best_math_schools, including "school_name" and "average_math" columns, sorted by "average_math" in descending order.

best_math_schools_results = schools[schools['average_math'] > (800*0.8)]
best_math_schools_columns = best_math_schools_results[['school_name','average_math']]
best_math_schools = best_math_schools_columns.sort_values(by='average_math',ascending=False)
best_math_schools.head(10)
```
